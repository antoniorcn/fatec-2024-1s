```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([150, 160, 155, 140, 170, 165, 152, 145, 168, 158, 143, 162, 165]) # metragens
```

```python
y = np.array([300000, 350000, 320000, 280000, 400000, 370000, 310000, 290000, 380000, 330000, 290000, 360000, 340000])
# preço das casas
```

```python
plt.scatter(x, y)
plt.title("Preços das Casas")
plt.xlabel("Tamanhos em m²")
plt.ylabel("Preços em reais R$")
plt.show()
```

```python
w = 2500
b = 0
y_hat = (x * w) + b
```

```python
y_hat
```

```python
plt.scatter(x, y, color="blue")
plt.plot(x, y_hat, color="red")
plt.title("Preços das Casas")
plt.xlabel("Tamanhos em m²")
plt.ylabel("Preços em reais R$")
plt.show()
```

```python
# Custo para w = 2800 e b = -110000
erro = y_hat - y
erro_2 = erro ** 2
erro_2
```

```python
soma = np.sum( erro_2 )
```

```python
soma
```

```python
m = x.shape[0]
custo = soma / (2 * m)
```
