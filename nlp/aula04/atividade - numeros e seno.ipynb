```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x = np.arange(0, 360, 1)
x = np.arange(0, 360, 1)
```

```python
x_rad = np.arange(0, 2 * np.pi, 0.1)
x_rad
```

```python
y = np.sin( x_rad )
```

```python
plt.plot(x_rad, y, color="red", linewidth=1, linestyle="dashed", 
         marker="*", markerfacecolor="blue", markersize=3)
plt.ylabel("Seno")
plt.xlabel("Angulo em Radianos")
plt.axvline(3.14)
plt.axhline(0)
plt.ylim(-2, 2)
plt.savefig("teste.png")
plt.show()
```

```python
plt.scatter(x_rad, y, color="red")
plt.ylabel("Seno")
plt.xlabel("Angulo em Radianos")
plt.axvline(3.14)
plt.axhline(0)
plt.ylim(-2, 2)
plt.show()
```
