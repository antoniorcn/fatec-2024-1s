```python
def derivada_wb(x, y, W, B ):
    m = len(x)
    soma_dw = 0
    soma_db = 0
    for i in range(m):
        valor_x = x[i]
        valor_y_real = y[i]
        valor_y_hat = valor_x * W + B
        dw_i = (valor_y_hat - valor_y_real) * valor_x
        db_i = (valor_y_hat - valor_y_real)
        soma_dw = soma_dw + dw_i
        soma_db = soma_db + db_i
    jdw_soma = soma_dw / m
    jdb_soma = soma_db / m
    return abs(jdw_soma), abs(jdb_soma)
```

```python
# Descida do Gradient para recalcular o W e o B
W = 1000
B = 1
taxa_aprendizagem = 0.000001
for era in range(616):
    custo_total_W, custo_total_B = derivada_wb(X, Y, W, B)
    print("Era:{}  Custo Total: {}".format(era, custo_total_W))
    W = W + taxa_aprendizagem * custo_total_W
    B = B + taxa_aprendizagem * custo_total_B
```
